```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Suicide rate.csv')
df.head()
```

```python
df.describe()
```

```python
#Data Cleaning

print(df.isnull().sum())
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
print(df.describe())
```

```python
df.drop('country-year', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Calculate correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix in a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Example: Histogram to show the distribution of suicides by age group
sns.histplot(df['age'], kde=True)
plt.title('Suicides by Age Group')
plt.show()
```

```python
# Example: Relationship between 'Age' and 'Year' in the DataFrame 'df'
sns.barplot(x='age', y='sex', data=df)
plt.title('Relationship between Age and Year')
plt.show()
```

```python
# Example: Relationship between 'column1' and 'column2' in the DataFrame 'df'
sns.scatterplot(x='age', y='suicides_no', data=df)
plt.title('Relationship between age and suicides_no')
plt.show()
```

```python
df.hist(bins=5, figsize=(10, 8),grid=False)
plt.show()
```
